# Modelling

```python
# Handling missing values
fil['Yield'].replace([np.inf, -np.inf], np.nan, inplace=True)
fil.dropna(subset=['Production', 'Yield'], inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

categorical_columns = ['State_Name','District_Name','Season', 'Crop','Months' ]  
for column in categorical_columns:
    fil[column] = label_encoder.fit_transform(fil[column])

```

```python
fil.drop(columns=['Crop_Year'], inplace=True)
```

```python
fil
```

```python
x= fil.drop('Yield',axis=1)
y= fil['Yield']
```

```python
# Remove infinite and large values
y.replace([np.inf, -np.inf], np.nan, inplace=True)
x.replace([np.inf, -np.inf], np.nan, inplace=True)

# Drop NaN values
x.dropna(inplace=True)
y.dropna(inplace=True)

# Ensure x and y have the same index
x = x.loc[y.index]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 1234)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# Feature Scaling

```python
from sklearn.feature_selection import SelectKBest, f_regression

selector = SelectKBest(score_func=f_regression, k=5)
X_train_selected = selector.fit_transform(x_train, y_train)
X_test_selected = selector.transform(x_test)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
# Linear Regression Model
regressor = LinearRegression()
model =regressor.fit(X_train_selected, y_train)
```

```python
# Make predictions
y_pred_lr = regressor.predict(X_test_selected)
```

```python
# Ensure y_test and y_pred_lr are of the same length
print("y_test shape:", y_test.shape)
print("y_pred_lr shape:", y_pred_lr.shape)
```

```python
regressor.coef_
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
# Evaluate the model
linear_reg_r2 = r2_score(y_test, y_pred_lr)
linear_reg_mse = mean_squared_error(y_test, y_pred_lr)
print("Linear Regression R2 score:", linear_reg_r2)
print("Linear Regression MSE:", linear_reg_mse)
```

```python
plt.scatter(y_test, y_pred_lr)
plt.title("Linear Regression Model")
plt.show()
```
